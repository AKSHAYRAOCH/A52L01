# Importing all the requried MODULE's

```python
  import numpy as np
  from matplotlib import pyplot as plt 
```

```python
b0 = np.array([1235, 1500, 1850, 2000])
b1 = np.array([1900, 1950, 2000, 2100])
x = np.array([3.2, 3.2, 3.7, 3.9, 4, 4, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6])
y = np.array([54445, 64445, 57189, 63218, 55794, 56957, 57081, 61111, 67938, 66029, 83088, 81363, 93940]) 
size = x.size 
```

## **Graph between Independent values and Dependent **

```python
plt.plot(x,y)
plt.title("Independent Values vs Dependent")
plt.xlabel(' X1 ')
plt.ylabel(' Y ')
plt.show()
```

## **The below code prints the Mean Squared Error for 4 different Beta values**

```python
for i in range(4): 
  Y_temp = []
  for j in range(size):
    temp = (b0[i] + (b1[i]*x[j])) ## This finds   " mX + c " value and stores it in temp variable
    Y_temp.append((y[j] - temp)**2 ) ## This find  ( y - y^ )^2 value and stores it in y_temp 
  print(f"Mean Square Error for Beta values {b0[i]} and {b1[i]} is -:-  {np.mean(Y_temp)}") ## This finds the Mean of the Squared " ( y - y^ )^2 " values!..
```
